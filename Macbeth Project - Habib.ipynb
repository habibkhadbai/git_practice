```python
### importing modules for the project
import requests
import matplotlib.pyplot as plt

### get data
macbeth = requests.get('http://www.gutenberg.org/cache/epub/2264/pg2264.txt').text

### displaying information about macbeth variable
print(type(macbeth))
print(len(macbeth))
print(macbeth[:500])

### split the text into words & unique words
word=macbeth.split()
unique_words = list(set(word))
print(unique_words)
len(word)
len(unique_words)

### creating a dictionary of unique words
uniquewordsdict = {}
for unique in unique_words:
    uniquewordsdict.update({unique : word.count(unique)})

### splitting the dictionary into unique words and counts
uniquewordslist=list(uniquewordsdict.keys())
print(uniquewordslist)
uniquewordscountlist=list(uniquewordsdict.values())
print(uniquewordscountlist)

### plotting graph size
plt.figure(figsize=(10,6), dpi=512)

### plotting the bar graph.
plt.bar(uniquewordslist, uniquewordscountlist, label='# of words')

### assigning labels
plt.xlabel('Unique Words',  rotation='vertical')
plt.ylabel('Word Count')
plt.title('Macbeth Word Count')
plt.legend()

### displaying the graph
plt.show()
```
